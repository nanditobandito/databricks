## Initialize default catalago and schema

```sql
use catalog workspace;
use schema test_schema;
```

## Create an empty delta table

```sql
create or replace table employees
(id INT, name STRING, salary DOUBLE)
```

## Fill in the employees table with some sample data
There are multiple insert statements in order to create multiple files

```sql
insert into employees 
values 
(1, 'George Washington', 1776),
(2, 'John Adams', 1802),
(3, 'Thomas Jefferson', 1809);

insert into employees
values 
(4, 'James Madison', 1817),
(5, 'James Monroe', 1825);

insert into employees
values
(6, 'John Quincy Adams', 1829),
(7, 'Andrew Jackson', 1837);

insert into employees
values
(8, 'Martin Van Buren', 1841),
(9, 'William Henry Harrison', 1841)
```

```sql
select
*
from employees
```

```sql
describe detail employees
```

## Lets give some of our founding fathers a pay bump

```sql
update employees 
set salary = salary * 1.776
where id <= 3
```
